## **Problem Description**
Predicting the age of abalone from physical measurements. The age of abalone is determined by cutting the shell through the cone, staining it, and counting the number of rings through a microscope -- a boring and time-consuming task. Other measurements, which are easier to obtain, are used to predict the age. Further information, such as weather patterns and location (hence food availability) may be required to solve the problem.

From the original data examples with missing values were removed (the majority having the predicted value missing), and the ranges of the continuous values have been scaled for use with an ANN (by dividing by 200).

```python
# Function to get the scatter plot of each feature against 'Age'

def get_scatter(df,X,y):
    import matplotlib.pyplot as plt
    %matplotlib inline
    import seaborn as sns
    print(f"{X} against {y}")
    plt.title(f"{X} with {y}")
    plt.scatter(data=df,x=X,y=y)
    plt.xlabel(X)
    plt.ylabel(y)
    plt.grid(True)
    return plt.show()


cols=['Length', 'Diameter', 'Height', 'Whole weight', 'Shucked weight',
       'Viscera weight', 'Shell weight']

col_iter = iter(cols)
y = 'age'
```

```python
X = next(col_iter,'List ended, no more elements')
get_scatter(df,X,y)
```

```python
### Logic to find outliers in Length
# length >=0.8 and age < 25
# length <=0.8 and age < 25


df[(df['Length'] >= 0.8) & (df['age'] <= 25)]
df[(df['Length'] < 0.8) & (df['age'] > 25)]

df.drop(df[(df['Length'] >= 0.8) & (df['age'] <= 25)].index,inplace=True)
df.drop(df[(df['Length'] < 0.8) & (df['Length'] > 25)].index,inplace=True)

```

### Selecting best model

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn import model_selection

models=[LinearRegression(),
          Ridge(),
          KernelRidge(alpha=1.0),
          SVR(),
          RandomForestRegressor(n_estimators=100),
          Lasso(),
          GradientBoostingRegressor(),
          KNeighborsRegressor(n_neighbors=2)
         ]

names=['ln_reg','ridge','k_ridge','svr','rf','ls_reg','gbr','knn']
results={}

for model,name in zip(models,names):
    kfold =  model_selection.KFold(n_splits=10)
    cv_results = model_selection.cross_val_score(model,x_train,y_train,cv=kfold)
    results[name] = cv_results
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)

# kfold = model_selection.KFold(n_splits=10)
# cv_results = model_selection.cross_val_score(LinearRegression()
#                                              ,x_train,y_train,cv=kfold)

```

```python
# results
df_results = pd.DataFrame(results)

df_means={}
df_stds={}
for name in names:
    df_means[name]=df_results[name].mean()
    df_stds[name]=df_results[name].std()
    

df_final=pd.concat([pd.DataFrame(names),pd.DataFrame(df_means.values()),pd.DataFrame(df_stds.values())],
                   axis=1
#                    ,names=(['model','mean','std'])
                  )
df_final.columns=['model','cv_mean','cv_std']
df_final.sort_values(by='cv_mean',ascending=False)
```

As per the above stats, Gradient boosting regressor and Random forest regressor are the best performing models, lets see the accuracy score below on test data

```python
def get_regression_eval(a,b):
    #### function to get rsuared, rsquared_adjusted, mse, rmse and mae
    ##### example >> get_regression_eval(y_test,_predict)
    import statsmodels.api as sm
    from sklearn.metrics import mean_squared_error,mean_absolute_error
    import math
    ######################################################################################
    get_stats={}
    print("############## Getting Rsquared and Adjusted Rsquared ########################")
    import statsmodels.api as sm
    
    a_addConstant = sm.add_constant(a)
    result = sm.OLS(b,a_addConstant).fit()
    get_stats['rsquared'] = result.rsquared
    get_stats['rsquared_adj'] = result.rsquared_adj
    
    print(get_stats)
    print("############## Getting MSE and RMSE ########################")
    get_stats['mse'] = mean_squared_error(y_test,_predict)
    get_stats['rmse'] = math.sqrt(mean_squared_error(y_test,_predict))
    
    print("############## Getting MAE(mean absolute error) ############")
    get_stats['mae'] = mean_absolute_error(y_test,_predict)
    print(get_stats)
    
    return get_stats

```

```python
### selecting top 3 models and see their accuracy on test data
### top 3 models based on the above analysis are >> Gradient Boosting Regression, Random Forest Regressor and Linear Regression.

from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor,RandomForestRegressor
import math

models = [GradientBoostingRegressor(),
          RandomForestRegressor(n_estimators=100),
          LinearRegression()]


models_alias=['gbr','rfr','lnr']

eval_dict={}
for model,name in zip(models,models_alias):
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    print(f"Evaluating score for {model}")
    eval_dict[name+'_score'] = get_regression_eval(y_test,y_pred)

    
    
eval_df = pd.DataFrame.from_dict(eval_dict)
```
